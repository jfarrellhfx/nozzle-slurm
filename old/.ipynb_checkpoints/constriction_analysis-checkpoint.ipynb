```python
results = "/scratch/alpine/jafa3629/constriction_1"
names = sorted(os.listdir(results))
print(names)
```

```python
q = np.load(f"{results}/V=11_910-eta=0_00100_npz.npz")['q']
q2 = np.load(f"{results}/V=11_960-eta=0_00100_npz.npz")['q']
```

```python
d = 307e-9
e0 = 8.854e-12
LY = 0.342
er = 3.9
C = e0*er/d
e = 1.602e-19
W = LY * 40e-6
m = 9.1e-31
V0 = -13
n0 = C*V0/(-e)
vs = np.sqrt(-e*V0/m)
I = 4
Res = 3000 / 4
J = -I / n0 / vs / 1000 / e / W
NY = 50

def VI(fname):
    q = np.load(f"{results}/{fname}")['q']
    V = 13 - (n0 * q[-1,0,:,NY//2]) * e / C
    I = n0 * vs * np.sum(q[-1,1,0,:])*1/50*0.342 * e * 40e-6*1000
    return V, I

V1, I1 = VI("V=11_910-eta=0_00100_npz.npz")
V2, I2 = VI("V=11_960-eta=0_00100_npz.npz")
dVdI1 = (V1 - V2) / (I1 - I2)

V11, I11 = VI("V=11_410-eta=0_00100_npz.npz")
V22, I22 = VI("V=11_460-eta=0_00100_npz.npz")
dVdI2 = (V11 - V22) / (I11 - I22)

V111, I111 = VI("V=11_010-eta=0_00100_npz.npz")
V222, I222 = VI("V=11_060-eta=0_00100_npz.npz")
dVdI3 = (V111 - V222) / (I111 - I222)
x = np.linspace(0,40,50)
```

```python
plt.plot(x[20:35], dVdI1[20:35],label = f"$I={I1:.3f}$")
plt.plot(x[20:35], dVdI2[20:35], label = f"$I={I11:.3f}$")
plt.plot(x[20:35], dVdI3[20:35], label = f"$I={I111:.3f}$")
plt.legend()
plt.xlabel("$x$ ($\mu$m")
plt.ylabel("$dV/dI$")
```

```python
results="/scratch/alpine/jafa3629/constriction_3"
Vs=[]
Is=[]
count = 0
names = os.listdir(results)[::10]
for name in names:
    q = np.load(f"{results}/{name}")['q']
    j = q[-1,1,-1,:]
    V = 13 - (n0 * q[-1,0,:,NY//2]) * e / C
    dV = np.abs(V[-1] - V[0])
    I = np.sum(j)*1/50*0.342 * e * 40e-6*1000 * n0 * vs
    print(f"{count}/{len(names)}")
    Vs.append(dV)
    Is.append(I)
    count += 1
```

```python
Is, Vs = zip(*sorted(zip(Is,Vs)))
Is = np.array(Is)
Vs = np.array(Vs)
dVdI = (Vs[1:]-Vs[:-1])/(Is[1:]-Is[:-1])
```

```python
results="/scratch/alpine/jafa3629/constriction_3"
print(sorted(os.listdir(results)[::5]))
```

```python
names = os.listdir(results)
Vs = []
for name in names:
    V = float(name[2:6].replace("_","."))
    Vs.append(V)
Vs, names = zip(*sorted(zip(Vs,names)))
print(names[::10])
plt.figure(figsize=(5,3.5))
x = np.linspace(0,40,100)
for name in names[:120][::20]:
    q = np.load(f"{results}/{name}")['q']
    j = q[-1,1,-1,:]
    V = 13 - (n0 * q[-1,0,:,NY//2]) * e / C
    I = np.sum(j)*1/50*0.342 * e * 40e-6*1000 * n0 * vs
    plt.plot(x[25:75], V[25:75], label = f"{I:.2f}")
plt.legend(title = "$I$ (mA)")
plt.xlabel("$x$ ($\mu$m)")
plt.ylabel("$V(x)$ (V)")

```

```python
plt.figure(figsize=(5,3.5))
for i, name in enumerate(names[:120]):
    if i % 20 == 0:
        q = np.load(f"{results}/{names[i]}")['q']
        q2 = np.load(f"{results}/{names[i+1]}")['q']
        j = q[-1,1,-1,:]
        j2 = q2[-1,1,-1,:]
        V = 13 - (n0 * q[-1,0,:,NY//2]) * e / C
        V2 = 13 - (n0 * q2[-1,0,:,NY//2]) * e / C
        I = np.sum(j)*1/50*0.342 * e * 40e-6*1000 * n0 * vs
        I2 = np.sum(j2)*1/50*0.342 * e * 40e-6*1000 * n0 * vs
        
        dVdI = (V-V2)/(I-I2)
        plt.plot(x[25:75], dVdI[25:75], label = f"{I:.2f}")
plt.legend(title = "$I$ (mA)")
plt.xlabel("$x$ ($\mu$m)")
plt.ylabel("$V(x)$ (V)")
```
